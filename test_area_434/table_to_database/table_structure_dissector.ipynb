```python
# Threshold for Empty cells (NAN's) in a dataframe. If the threshold is surpassed, the row/column is regarded as a DUPLICATE
THRESHOLD = 0.1 # Threshold is taken as a percentage value.
```

```python
# (7,1 : 8, 1)
# This function checks if the cells below are empty
def contain_duplicate_rows(df, cell_1, cell_2):
    print("dup rows", cell_1, cell_2)
    
    # +1 because python range exludes the last element
    c_df = df.iloc[cell_1[0]:cell_2[0]+1, cell_1[1]:cell_2[1]+1]
    
    # Go through each row of the "rectangle" (extracted subtable)
    for i in range(len(c_df)):
        row = c_df.iloc[i]
        # An entire row is nan
        if row.isnull().values.sum() == len(row) :
            return True
        
    return False
    
    
# This function checks if the cells to right are empty
def contain_duplicate_cols(df, cell_1, cell_2):
    print("dup cols", cell_1, cell_2)
    # Edge case (only in the begining)
    if cell_1 == cell_2:
        return False
    # Get the rectangle
    c_df = df.iloc[cell_1[0]:cell_2[0]+1, cell_1[1]:cell_2[1]+1]
    last_row = c_df.iloc[-1]
    
    return last_row.isnull().values.sum() == len(last_row)
    
```

```python
# Note this code was lifted from the paper
def get_cc1_and_cc2(df, cc4):
    # Initialise the user-defined last data cell
    c_max = cc4[1]
    r_max = cc4[0]
    
    # Initalise the starting points
    r1 = c1 = 1
    r2 = r_max - 1
    c2 = 1
    
    # Initalise the flags as well as the area variable (the coords with the maximum area will be the resulting MIP)
    right_flag = up_flag = 0
    max_area = 0
    
    # Prep the sub_header returning points
    cc1 = [r1, c1]
    cc2 = [r2, c2]
    
    while c2 < c_max and r2 >= r1:
        print(c1, c2, r1, r2)
        if not contain_duplicate_rows(df, cell_1=[r2+1, c1], cell_2=[r_max, c2]) and not contain_duplicate_cols(df, cell_1=[r1, c2+1], cell_2=[r2-1, c_max]):
#             print("No row duplicates")
#             if not contain_duplicate_cols(df, cell_1=[r1, c2+1], cell_2=[r2-1, c_max]):
#                 print("No Col duplicates")
            r2 -= 1 
            up_flag = 1
            right_flag = 0
        else:
            c2 += 1
            right_flag += 1
            print("in else")
            if up_flag == right_flag: # Both should be equal to 1 
                data_area = (r_max - r2 + 1) * (c_max - c2 + 1)
                if data_area > max_area:
                    max_area = data_area
                    cc2 = [r2, c2]
                up_flag = 0
                
        print(cc2)
                


    # Locate CC1 at intersection of the top row and the leftmost column necessary for indexing
    r1 = c1 = 1
    print("cc2", cc2)
    print("exit loop")
    while not contain_duplicate_cols(df, cell_1=[r1+1, c2+1], cell_2=[r2, c_max]):
        r1 += 1
    while not contain_duplicate_rows(df, cell_1=[r2+1, c1+1], cell_2=[c2, r_max]):
        c1 += 1

    cc1 = [r1, c1]

    return cc1, cc2   
```
